```python
import matplotlib.pyplot as plt

# Data for different countries
countries_data = {
    'India': [3.75, 7.70, 3.60, 2.94, 2.49, 2.40],
    'Brazil': [1.28,5.58,7.72,0.96,1.64,0.96],
    'Russia':[5.53,10.14,8.40,4.03,3.10,2.89],
    'US': [3.57,6.99,7.96,3.24,5.05,2.75]
}

# Categories
categories = ['ARIMA', 'Prophet', 'LSTM', 'GRU', 'ARIMA+ANN', 'Proposed ARIMA+LSTM']

# Plotting
plt.figure(figsize=(10, 6))

# Plotting for each country
for country, values in countries_data.items():
    plt.plot(categories, values, marker='o', label=country)

# Adding labels and title
plt.xlabel('Models')
plt.ylabel('Values')
plt.title('Comparison of MAPE for Different Countries')

# Adding legend
plt.legend()

# Displaying the plot
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
